# D. Group & Visualize the data

```python
grouped = df.groupby("CSA")["D5DE"].mean() #Sorting "DFDE" (Employment Density) by "CSA" (Combined statistical area) 
```

```python
grouped.head()
```

```python
import matplotlib.pyplot as plt

# --- Chart 1: Employment Density by CSA ---
grouped.head().plot(
    kind="bar",
    x="CSA",
    y="D5DE",
    legend=False
)

plt.title("Employment Density (D5DE) by CSA")
plt.ylabel("D5DE (Employment Density, transformed)")
plt.xlabel("CSA")
plt.xticks(rotation=0)  # keep CSA codes horizontal
plt.tight_layout()
plt.show()



```

```python
import matplotlib.pyplot as plt

sum_walkability_by_CSAName = df.groupby('CSA_Name')['NatWalkInd'].sum()
sum_walkability_by_CSAName.nlargest(10).plot.bar()

plt.title("Top 10 Metropolitan Areas by Walkability (NatWalkInd Sum)")
plt.ylabel("Total Walkability Index (NatWalkInd)")
plt.xlabel("Metropolitan Area (CSA_Name)")
plt.xticks(rotation=45, ha="right")  # makes x-axis labels easier to read

plt.show()
```

I grouped employment data and walkability based on CSA (Location Name). I then created to charts to show the results.
